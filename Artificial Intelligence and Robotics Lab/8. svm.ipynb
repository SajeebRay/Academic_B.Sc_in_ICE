```python
df=pd.read_csv('Social_Network_Ads.csv')
print(df.head())
```

```python
x=df.iloc[:,[0,1]]
x.head()
```

```python
y=df.iloc[:,2]
y.head()
```

```python
print('Training data : ',xtrain.shape)
print('Testing data : ',xtest.shape)
xtrain.head()
```
